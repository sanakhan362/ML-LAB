## Handling Missing Values

```python
#Replace the msiing values in Embarked column by 'S'
import pandas as pd

# Load the dataset
df = pd.read_csv("train_titanic.csv")  # Load the correct file

# Replace missing values in 'Embarked' column with 'S'
df["Embarked"].fillna("S", inplace=True)

# Verify if missing values are replaced
print(df["Embarked"].isnull().sum())  # Should print 0 if all missing values are replaced
```

```python
#Impute the missing values in Fare column with median value
import pandas as pd

# Load the dataset
df = pd.read_csv("train_titanic.csv")  # Load the correct file

# Replace missing values in 'Fare' column with its median value
df["Fare"].fillna(df["Fare"].median(), inplace=True)

# Verify if missing values are replaced
print(df["Fare"].isnull().sum())  # Should print 0 if all missing values are replaced
```

```python
#Impute the Age column. It has too many missing values but it may be important. 
#We fill the missing values with numbers randomly drawn from distribution
#that has same mean and standard deviation as non missing Age data
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv("train_titanic.csv")  # Load the correct file

# Calculate mean and standard deviation of the non-missing Age values
age_mean = df["Age"].mean()
age_std = df["Age"].std()

# Count missing values in Age column
num_missing = df["Age"].isnull().sum()

# Generate random values from a normal distribution with the same mean and std dev
random_ages = np.random.normal(loc=age_mean, scale=age_std, size=num_missing)

# Replace missing Age values with the generated random values
df.loc[df["Age"].isnull(), "Age"] = random_ages

# Verify if missing values are replaced
print(df["Age"].isnull().sum())  # Should print 0 if all missing values are replaced
```

```python
# Splitting the data into train and validation set - 90% train 10% validation
#X_train, X_test, y_train, y_validation
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Load the dataset
df = pd.read_csv("train_titanic.csv")  # Load the correct file

# Drop unnecessary columns
df.drop(columns=["PassengerId", "Name", "Ticket", "Cabin"], inplace=True)

# Identify categorical columns
categorical_columns = ["Sex", "Embarked"]

# Apply Label Encoding to categorical features
df_categorical = df[categorical_columns].apply(LabelEncoder().fit_transform)

# Drop original categorical columns
df = df.drop(columns=categorical_columns)

# Concatenate encoded categorical features with the original DataFrame
df = pd.concat([df, df_categorical], axis=1)

# Prepare X and y
X = df.drop(columns=["Survived"])  # Features
y = df["Survived"]  # Target variable

# Splitting the data into 90% train and 10% validation sets
X_train, X_validation, y_train, y_validation = train_test_split(X, y, test_size=0.1, random_state=42)

# Display the shapes of the sets
print(f"X_train shape: {X_train.shape}")
print(f"X_validation shape: {X_validation.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_validation shape: {y_validation.shape}")
```

```python
# Create the model
# Fitting the decision tree with default hyperparameters
# Importing decision tree classifier from sklearn library
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Load the dataset
df = pd.read_csv("train_titanic.csv")  # Load the correct file

# Drop unnecessary columns
df.drop(columns=["PassengerId", "Name", "Ticket", "Cabin"], inplace=True)

# Identify categorical columns
categorical_columns = ["Sex", "Embarked"]

# Apply Label Encoding to categorical features
df_categorical = df[categorical_columns].apply(LabelEncoder().fit_transform)

# Drop original categorical columns from df
df = df.drop(columns=categorical_columns)

# Concatenate encoded categorical features with the original DataFrame
df = pd.concat([df, df_categorical], axis=1)

# Define target variable (y) and feature set (X)
y = df["Survived"]
X = df.drop(columns=["Survived"])

# Split the data into train (90%) and validation (10%) sets
X_train, X_validation, y_train, y_validation = train_test_split(X, y, test_size=0.1, random_state=42)

# Create Decision Tree model with default hyperparameters
model = DecisionTreeClassifier(random_state=42)

# Fit the model on the training data
model.fit(X_train, y_train)

# Predict on the training set
y_train_pred = model.predict(X_train)

# Predict on the validation set
y_validation_pred = model.predict(X_validation)

# Compute accuracy scores
train_accuracy = accuracy_score(y_train, y_train_pred)
validation_accuracy = accuracy_score(y_validation, y_validation_pred)

# Print accuracy
print(f"Training Accuracy: {train_accuracy:.4f}")
print(f"Validation Accuracy: {validation_accuracy:.4f}")
```

```python
#Print the tree

# Importing required packages for visualization
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# Set figure size for better readability
plt.figure(figsize=(15, 8))

# Plot the decision tree
plot_tree(model, filled=True, feature_names=X.columns, class_names=["Not Survived", "Survived"], rounded=True)

# Show the plot
plt.show()

```

```python
# plotting tree with max_depth=3
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Create a Decision Tree model with max_depth=3
model_pruned = DecisionTreeClassifier(max_depth=3, random_state=42)

# Fit the model on the training data
model_pruned.fit(X_train, y_train)

# Set figure size
plt.figure(figsize=(15, 8))

# Plot the decision tree with max_depth=3
plot_tree(model_pruned, filled=True, feature_names=X.columns, class_names=["Not Survived", "Survived"], rounded=True)

# Show the plot
plt.show()
```
