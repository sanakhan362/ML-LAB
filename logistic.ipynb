```python
X=df.drop(columns="Species")
y=df["Species"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
print(X_train.shape)
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print(y_pred)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Species']=le.fit_transform(df['Species'])
df
```

```python
model=SVC(C=1)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report,recall_score,precision_score,f1_score
print(confusion_matrix(y_pred,y_test))
```

```python
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
print(recall_score(y_pred,y_test,average='macro'))
print(f1_score(y_pred,y_test,average='macro'))
print(precision_score(y_pred,y_test,average='macro'))
```

```python
from sklearn.model_selection import KFold,cross_val_score
kf=KFold(n_splits=5,shuffle=True,random_state=1)  #split dataset into 5 folds
cv=cross_val_score(model,X,y,cv=kf)# shuffle the data before split into folds
print("accuracy of each fold",cv)
print("mean accuracy:",cv.mean())
```

```python
from sklearn.model_selection import GridSearchCV
param_g={'C':[0.1,1,10,100,1000]}      #
grid=GridSearchCV(SVC(),param_g,cv=5,return_train_score=True)
grid.fit(X,y)
```

```python
result=pd.DataFrame(grid.cv_results_)
print(result[['param_C','mean_test_score','rank_test_score']])
```

```python
import matplotlib.pyplot as plt
C_values=result['param_C'].astype(float)
train_score=result['mean_train_score']
test_score=result['rank_test_score']
plt.plot(C_values,train_score,label='Train score',marker='o')
plt.plot(C_values,test_score,label='test_score',marker='o')
plt.xlabel('C value')
plt.ylabel('Accuracy')
plt.title('Train and test score for different C values')
plt.legend()
plt.grid(True)
plt.show()
```
